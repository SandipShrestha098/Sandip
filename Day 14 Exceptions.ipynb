```python
addition(3,4,5,6)
```

```python
def addition(*args): #variable no. of arguments
    add = 0
    for arg in args:
        add += arg
    return add
```

```python
addition(3,4,5,6)
```

```python
def addition(**kwargs): #variable no. of keyword arguments
    print(kwargs)
    print(type(kwargs))
```

```python
addition(a=1, b=2)
```

```python
def addition(**kwargs): #variable no. of arguments
    add = 0
    for value in kwargs.values():
        add += value
    return add
```

```python
addition(a=1, b=2)
```

```python
addition(3,4,5, a=1, b=2)
```

```python
def addition(*args, **kwargs):
    print(args, kwargs)
```

```python
addition(3,4,5, a=1, b=2)
```

```python
def addition(*args, **kwargs): #variable no. of arguments
    add = 0
    for arg in args:
        add +=arg
        
    for value in kwargs.values():
        add += value
    return add
```

```python
addition(3,4,5, a=1, b=2)
```

```python
import json

person = {
    "name": "Sandip",
    "country" : "Nepal",
    "city" : "Kathmandu",
    'skills': [ 
        "Python",
        "MATLAB",
        "R"
    ],
    "x" : 123,
}

print (person, type (person))

person_json = json.dumps(person)
print (person_json, type(person_json))
```
